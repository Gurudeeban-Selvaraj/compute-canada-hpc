```python
# Import the Excel file
dataset = pd.read_excel("cid_smiles.xlsx")
```

```python
dataset
```

```python
X = dataset.drop('CID', axis=1)
X
```

```python
qeds = []  # initialize an empty list to store the QED values
for index, smiles in X.iterrows():
    mol = Chem.MolFromSmiles(smiles['smiles'])
    qeds.append(mol) # Loop over the SMILES and convert them to molecules
   
# Calculate QED values
qed_values = [qed.default(mol) for mol in qeds] 
print(qed_values)

# write the DataFrame to a CSV file
X['QED'] = qed_values
X.to_csv('92mol_qed.csv', index=False)
```

```python
qeds = []  # initialize an empty list to store the QED values
for index, smiles in X.iterrows():
    mol = Chem.MolFromSmiles(smiles['smiles'])
    p = qed.properties(mol)
    qeds.append(p)  # add the QED value to the list
    print(p)
    
# write the DataFrame to a CSV file
qeds_df = pd.DataFrame(qeds, columns=['MW', 'ALOGP','HBA', 'HBD', 'PSA', 'ROTB','AROM','ALERTS'])
qeds_df.to_csv('92mol_qeds_properties.csv', index=False)
```
